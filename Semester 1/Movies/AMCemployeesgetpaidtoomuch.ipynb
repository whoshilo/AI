```python
def get_best_fit(data, x, y):
    '''
    Determines the slope (m) and intercept (b) of the best fit line
    for two variables, x and y.
    '''
    m = 0
    for count in range(100):
        m = m + 1
        b = 0

        # Calculate residuals (y = mx+b)
        res = 0
        for i in range(len(data)):
            pred_y = (m * data.loc[i][x]) + b
            act_y = data.loc[i][y]
            temp = act_y - pred_y 
            temp = temp ** 2
            res = res + temp
        print(res)
    return (m, b, res)
    #1.26x + 1048
```

```python
def CoeDet(df, x, y,m,b):
    y_mean = df[y].mean()
    Tsst = 0
    for value in df[y]:
        diff = (value - y_mean)
        squ = diff ** 2
        Tsst += squ

    Tssr = 0

    for i in range(len(df)):
        predicted = (m * df.loc[i][x]) + b
        diff = predicted - df.loc[i][y]
        squ = diff ** 2
        Tssr += squ 

    r2 = 1 - (Tssr / Tsst)
    return r2
```
