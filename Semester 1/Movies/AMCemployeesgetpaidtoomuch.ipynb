```python
def get_best_fit(data, x, y):
    '''
    Determines the slope (m) and intercept (b) of the best fit line
    for two variables, x and y.
    '''
    m = 0
    for count in range(100):
        m = m + 1
        b = 0

        # Calculate residuals (y = mx+b)
        res = 0
        for i in range(len(data)):
            pred_y = (m * data.loc[i][x]) + b
            act_y = data.loc[i][y]
            temp = act_y - pred_y 
            temp = temp ** 2
            res = res + temp
        print(res)
    return (m, b, res)
    #1.78x + 1279
```

```python
def CoeDet(df, x, y,m,b):
    x_mean = df[x].mean()
    y_mean = df[y].mean()

    num = ((df[x] - x_mean) * (df['y'] - y_mean)).sum()
    den = ((df[x] - x_mean) **2).sum()
    slope = num/den

    int = y_mean - slope * x_mean

    return slope, int
```
